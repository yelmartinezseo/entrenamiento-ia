```python
import gym
import numpy as np
import random

class MarsExplorerEnv(gym.Env):
    metadata = {'render.modes': ['human']}

    def __init__(self):
        super().__init__()
        self.action_space = gym.spaces.Discrete(2)  # 0: izquierda, 1: derecha
        self.observation_space = gym.spaces.Discrete(10)  # Estados de 0 a 9
        self.state = 5
        self.interest_zone = 9
        self.danger_zones = [2, 4, 7]

    def step(self, action):
        """Aplica la acción y devuelve (estado, recompensa, terminado, truncado, info)."""
        if action == 0:
            self.state = max(0, self.state - 1)
        else:
            self.state = min(self.observation_space.n - 1, self.state + 1)

        terminated = self.state == self.interest_zone
        reward = 10 if terminated else (-5 if self.state in self.danger_zones else 0)

        truncated = False  # No hay límite de pasos en este entorno.
        info = {}

        return self.state, reward, terminated, truncated, info

    def reset(self, seed=None, options=None):
        """Reinicia el entorno y devuelve (estado, info)."""
        super().reset(seed=seed)
        self.state = 5
        info = {}
        return self.state, info

    def render(self, mode='human'):
        """Muestra la posición del explorador en la línea de estados."""
        if mode != 'human':
            raise NotImplementedError("Solo se admite el modo 'human'.")

        line = ['.'] * 10
        line[self.state] = '🚀'  # Representación del explorador
        print("".join(line))  # Visualización

    def seed(self, seed=None):
        """Establece la semilla para la reproducibilidad."""
        np.random.seed(seed)

# Crear y ejecutar el entorno
env = MarsExplorerEnv()

episodes = 10
for episode in range(1, episodes + 1):
    state, _ = env.reset()
    done = False
    total_reward = 0

    print(f"\n🎬 Episodio {episode}")

    while not done:
        action = env.action_space.sample()
        state, reward, done, _, _ = env.step(action)
        total_reward += reward
        env.render()

    print(f"🏆 Recompensa total: {total_reward}")
```
